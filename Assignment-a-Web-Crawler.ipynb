```python
#---
# Importing all the require libraries and packages for Python to run for this assignment:
# - Requests: It allows the developer to send HTTP/1.1 requests using Python. With it, you can add content 
#             like headers, form data, multipart files, and parameters via simple Python libraries. 
#             It also allows you to access the response data of Python in the same way.
# - Pandas: The package comes with several data structures that can be used for many different data manipulation tasks. 
#           It also has a variety of methods that can be invoked for data analysis, which comes in handy when working 
#           on data science and machine learning problems in Python.
# - Numpy: It is a general-purpose array-processing package. It provides a high-performance multidimensional array object, 
#          and tools for working with these arrays.
# - BeautifulSoup: Is a library for pulling data out of HTML and XML files. It works with your favorite parser to provide 
#                  idiomatic ways of navigating, searching, and modifying the parse tree. It commonly saves programmers 
#                  hours or days of work.
# - Tabulate: Is smart about column alignment. It detects columns which contain only numbers, and aligns them by a decimal 
#             point (or flushes them to the right if they appear to be integers).
#---
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from tabulate import tabulate
```

```python
### ======================================== ###
### NO. OF CONFIRMED CASES
### ======================================== ###
```

```python
url = "https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"

try:
    page = requests.get(url, timeout=5)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content,'html.parser')
        table = soup.find("table", {"class": "js-csv-data csv-data js-file-line-container"})
        df = pd.read_html(str(table))
        print(df)
    else: 
        print(str(page.status_code) + " - Error, page not found.")
except requests.ConnectionError as e:
    print('Connection error')
    print(str(e))
```

```python
data = df[0]
```

```python
print(type(data))
```

```python
print(data.describe())
```

```python
data.describe().transpose()
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data = data.drop(['Lat','Long'], axis=1)
```

```python
# Get data for ASEAN regions:
asean = ['Malaysia', 'Singapore', 'Thailand', 'Brunei', 'Cambodia', 'Indonesia', 'Laos', 'Myanmar', 'Philippines', 'Vietnam']
```

```python
asean_countries = data['Country/Region'].isin(asean)
```

```python
data_asean = data[asean_countries]
```

```python
data_asean = data_asean.drop('Unnamed: 0', axis=1)
```

```python
data_asean = data_asean.drop(['Lat','Long'], axis=1)
```

```python
data_asean = data_asean.drop('Province/State', axis=1)
```

```python
data_asean = data_asean.rename(columns={'Country/Region':'Country'})
```

```python
data_asean
```

```python
data_asean = data_asean.set_index('Country')
```

```python
data_asean
```

```python
data_asean_t = data_asean.transpose()
```

```python
url_recovered = "https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv"

try:
    page = requests.get(url_recovered, timeout=5)
    if page.status_code == 200:
        soup = BeautifulSoup(page.content,'html.parser')
        table = soup.find("table", {"class": "js-csv-data csv-data js-file-line-container"})
        df_recovered = pd.read_html(str(table))
#         print(df)
    else: 
        print(str(page.status_code) + " - Error, page not found.")
except requests.ConnectionError as e:
    print('Connection error')
    print(str(e))
```

```python
df_recovered = df_recovered[0]
```

```python
df_recovered.head()
```

```python
asean_recovered = df_recovered.drop('Unnamed: 0', axis=1)
asean_recovered = asean_recovered.drop(['Lat','Long'], axis=1)
asean_recovered = asean_recovered.drop('Province/State', axis=1)
asean_recovered = asean_recovered.rename(columns={'Country/Region':'Country'})
asean_recovered = asean_recovered.set_index('Country')
asean_recovered
```

```python
select_asean = asean_recovered.index.isin(asean)
asean_recovered = asean_recovered[select_asean]
asean_recovered_t = asean_recovered.transpose()
asean_recovered_t
```

```python
asean_recovered_t.plot()
plt.ylabel('No. of recovered cases')
plt.xlabel('Time')
plt.show()
```

```python
# COMMENT: Singapore is leading in the number of covid-19 recovery.
# Followed by Thailand and Malaysia.
```

```python
### ======================================== ###
### MAINLAND CHINA (GROUPED)
### ======================================== ###
confirmed_cases_china = data.drop('Unnamed: 0', axis=1)
confirmed_cases_china = confirmed_cases_china.drop(['Lat','Long'], axis=1)
confirmed_cases_china = confirmed_cases_china.drop('Province/State', axis=1)
confirmed_cases_china = confirmed_cases_china.rename(columns={'Country/Region':'Country'})
confirmed_cases_china = confirmed_cases_china.set_index('Country')
confirmed_cases_china
```

```python
select_china = confirmed_cases_china.index == 'Mainland China'
confirmed_cases_china = confirmed_cases_china[select_china]
```

```python
confirmed_cases_china
```

```python
china_combined_confirmed = confirmed_cases_china.groupby('Country').sum()
```

```python
china_combined_confirmed
```

```python
china_combined_confirmed_t = china_combined_confirmed.transpose()
china_combined_confirmed_t
```

```python
china_combined_confirmed_t.plot()
```

```python
# COMMENT: Confirmed cases in China continues to rise. More than 80k cases confirmed.
```
